```python
import pandas as pd
import logging
from seq2seq_model import Seq2SeqModel
logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)
```

# Data Preperation

```python
train_data = pd.read_csv("./data/conll2003/train.csv", sep=',').values.tolist()
train_df = pd.DataFrame(train_data, columns=["input_text", "target_text"])
train_subset = train_df[:100]
```

```python
eval_data = pd.read_csv("./data/conll2003/dev.csv", sep=',').values.tolist()
eval_df = pd.DataFrame(eval_data, columns=["input_text", "target_text"])
eval_subset = eval_df[:25]
```

# Modelling

```python
model_args = {
    "reprocess_input_data": True,
    "overwrite_output_dir": True,
    "max_seq_length": 50,           # Original : 50
    "train_batch_size": 10,          # Original : 100
    "num_train_epochs": 10,         # Changeable | Fina change to 10 | Original : 20
    "save_eval_checkpoints": False,
    "save_model_every_epoch": False,
    "evaluate_during_training": True,
    "evaluate_generated_text": True,
    "evaluate_during_training_verbose": True,
    "use_multiprocessing": False,
    "max_length": 25,
    "manual_seed": 4,
    "save_steps": 11898,
    "gradient_accumulation_steps": 1,
    "output_dir": "./exp/template",
}
```

```python
# Initialize model
model = Seq2SeqModel(
    encoder_decoder_type="bart",            # Bisa diganti2
    encoder_decoder_name="facebook/bart-large",     # Bisa diganti2
    args=model_args,
    use_cuda=False,     # Source code : comment | Alfina : uncomment
                            # Uncomment if train locally, comment if train on GCP
)
```

# Training

```python
# Train the model
model.train_model(train_subset, eval_data=eval_subset)
```

# Model Eval

```python
# Evaluate the model
results = model.eval_model(eval_subset)
```

```python
# Use the model for prediction

print(model.predict(["Japan began the defence of their Asian Cup title with a lucky 2-1 win against Syria in a Group C championship match on Friday."]))
```
